```python
X = np.arange(1,101)
for model_number in range(1,9):
  loss = pd.read_csv(file_path + '2 Loss/' + 'loss_'+str(model_number)+'.csv', usecols = ['0'])
  plt.scatter(X, loss, label='model '+str(model_number), s=5)

plt.xlabel("epoch",fontsize=18)
plt.ylabel("loss",fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14, loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
X = np.arange(1,101)
for model_number in range(1,9):
  val_loss = pd.read_csv(file_path + '3 Validation loss/' + 'val_loss_'+str(model_number)+'.csv', usecols = ['0'])
  plt.plot(X, val_loss, label='model'+str(model_number))

plt.xlabel("epoch")
plt.ylabel("validation loss")
plt.legend()
```

```python
X = np.arange(1,101)
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

for model_number in range(1,9):
  loss = pd.read_csv(file_path + '2 Loss/' + 'loss_'+str(model_number)+'.csv', usecols = ['0'])
  val_loss = pd.read_csv(file_path + '3 Validation loss/' + 'val_loss_'+str(model_number)+'.csv', usecols = ['0'])

  ax[0].scatter(X, loss, label='model '+str(model_number), s=5)
  ax[0].set_xlabel('epoch',fontsize=16)
  ax[0].set_ylabel('loss',fontsize=16)

  ax[1].scatter(X, val_loss, label='model '+str(model_number), s=5)
  ax[1].set_xlabel('epoch',fontsize=16)
  ax[1].set_ylabel('validation loss',fontsize=16)

for axis in ax:
    axis.tick_params(axis='both', labelsize=14)

plt.subplots_adjust(wspace=0.4)
plt.legend(fontsize=14, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
